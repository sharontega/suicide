```python
num=df[['year','suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)',' gdp_for_year ($) ']]
cal=df[['sex','age','generation']]
```

```python
for i in num:
    sns.histplot(num[i])
    plt.title(i)
    plt.show()
```

```python
for i in cal:
    sns.countplot(cal[i])
    plt.title(i)
    plt.show()
```

```python
df.rename(columns={' gdp_for_year ($) ':'gdp_for_year', 'gdp_per_capita ($)':'gdp_per_capita'},inplace=True)
```

```python
df.columns=df.columns.str.replace(' ','')
```

```python
df['gdp_for_year']=df.gdp_for_year.str.replace(',','')
```

```python
df['gdp_for_year']=df.gdp_for_year.astype('float')
```

```python
cor=df.select_dtypes(py.number).corr()
sns.heatmap(cor,annot=True)
```

```python
sns.pointplot(df.generation,df.suicides_no)
```

```python
sns.lineplot(df.age,df.suicides_no)
```

```python
sns.lineplot(df.year,df.suicides_no)
```

```python
s=df.groupby('country').sum().reset_index().sort_values('suicides_no',ascending=False)
plt.figure(figsize=(6,32))
sns.barplot(x='suicides_no',y='country',data=s)
a=plt.xticks(rotation=90)
plt.title('country by suicide_no')
plt.show()
```
